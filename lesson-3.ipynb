```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, NMF
from sklearn.preprocessing import MinMaxScaler
from scipy import signal

# Step 1: Generate synthetic signals
rng = np.random.RandomState(0)
n_samples = 2000
time = np.linspace(0, 8, n_samples)

s1 = np.sin(2 * time)                   # Sinusoidal signal
s2 = np.sign(np.sin(3 * time))         # Square signal
s3 = signal.sawtooth(2 * np.pi * time) # Sawtooth signal

S = np.c_[s1, s2, s3]
S += 0.2 * rng.normal(size=S.shape)  # Add noise
S /= S.std(axis=0)                   # Standardize

# Step 2: Mix signals to create observations
A = rng.randn(3, 3)  # Mixing matrix
X = S @ A.T          # Observed signals (mixed)

# Step 3: Apply PCA
pca = PCA(n_components=3)
H = pca.fit_transform(X)

# Step 4: Apply NMF (requires non-negative input)
scaler = MinMaxScaler()
X_nmf_input = scaler.fit_transform(X)  # Scale to [0, 1]
nmf = NMF(n_components=3, init='random', random_state=0, max_iter=1000)
S_ = nmf.fit_transform(X_nmf_input)

# Step 5: Plot original, mixed, and recovered signals
models = [X, S, S_, H]
names = ['Observations (mixed signals)', 'True sources (original)',
         'NMF recovered sources', 'PCA recovered sources']

fig, axes = plt.subplots(4, 1, figsize=(10, 8), sharex=True)
for ax, model, name in zip(axes, models, names):
    ax.plot(model)
    ax.set_title(name)
plt.tight_layout()
plt.show()
```
